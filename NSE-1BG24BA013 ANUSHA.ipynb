```python
| Column         | Description                                       |
| -------------- | ------------------------------------------------- |
| `show_id`      | Unique ID for each Netflix title                  |
| `type`         | Indicates whether the title is a Movie or TV Show |
| `title`        | Name of the movie or TV show                      |
| `director`     | Director of the content                           |
| `country`      | Country where the content was produced            |
| `release_year` | Year the content was released                     |
```

```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("netflix_titles.csv")
df
```

```python
# Shape of dataset
df.shape
```

```python
# First 5 rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
#Missing Values & Duplicate Check
# Missing values count
df.isnull().sum()
```

```python
# Duplicate rows
df.duplicated().sum()
```

```python
#Data Cleaning
# 1. Remove duplicate rows
df = df.drop_duplicates()
df
```

```python
# 2. Convert date_added to datetime
df['date_added'] = pd.to_datetime(
    df['date_added'],
    format='%B %d, %Y',
    errors='coerce'
)
```

```python
# 3. Fill missing director with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df
```

```python
# 4. Fill missing country with mode
df['country'] = df['country'].fillna(df['country'].mode()[0])
df
```

```python
# 5. Strip extra spaces from text columns
df['type'] = df['type'].str.strip()
df['rating'] = df['rating'].str.strip()
df
```

```python
# 6. Convert type column to lowercase
df['type'] = df['type'].str.lower()
df
```

```python
# 7. Standardize country casing
df['country'] = df['country'].str.title()
df
```

```python
#Create 3 New Columns
# 1. Content Age
df['content_age'] = 2025 - df['release_year']
df
```

```python
# 2. Is Recent Content (Conditional)
df['is_recent'] = np.where(df['release_year'] >= 2018, 'Yes', 'No')
df
```

```python
# 3. Year Added to Netflix
df['year_added'] = df['date_added'].dt.year
df
```

```python
#Generate 5 Insights
#Insight 1: Count of Movies vs TV Shows(value_counts)
df['type'].value_counts()
```

```python
#Insight 2: Top 10 Content Producing Countries(groupby + agg)
df.groupby('country')['show_id'].count().sort_values(ascending=False).head(10)
```

```python
#Insight 3: Oldest Content on Netflix(sort_values)
df.sort_values(by='release_year').head(5)
```

```python
#Insight 4: Movies vs TV Shows by Year Added(pivot_table)
pd.pivot_table(
    df,
    values='show_id',
    index='year_added',
    columns='type',
    aggfunc='count'
)
```

```python
#Insight 5: Recent Movies Only(Boolean filtering)
df[(df['type'] == 'movie') & (df['is_recent'] == 'Yes')]
```
