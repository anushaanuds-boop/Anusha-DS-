```python
# Student Record Management System

## Project Overview
The Student Record Management System is a menu-driven Python mini project designed to manage student information efficiently.
It allows users to add, view, search, and delete student records using a simple console-based interface.
The project demonstrates the practical application of core Python concepts such as data structures, functions, loops, exception handling, and file handling.

## Problem Statement
Managing student records manually can lead to errors, data loss, and difficulty in searching or updating records.
This project provides a simple digital solution to store and manage student details in an organized and efficient manner.

## Features
- View all student records  
- Add new student records  
- Search student details using Student ID  
- Delete student records  
- Menu-driven program for easy interaction  
- Data persistence using file handling  

## Data Fields Used
- Student ID  
- Student Name  
- Course  
- Semester  
- Marks  

## Python Concepts Used

### Data Structures
- Dictionary of Lists – Used to store student details  
- Lists – Used to store multiple values for each student attribute  

### Libraries Used
- pandas  
- numpy  

### Control Structures
- while loop  
- if / elif / else statements  

### Functions Used
- view_students() – Displays all student records  
- add_student() – Adds a new student record  
- search_student() – Searches a student using ID  
- delete_student() – Deletes a student record  
- save_to_file() – Saves student data into a file  

### Methods Used
- List methods: append(), pop(), len()  
- Dictionary methods: keys(), values(), get()  
- String methods: strip(), title()  

### Exception Handling
- try / except blocks are used to handle invalid inputs such as entering text instead of numbers.

### File Handling
- Student data is stored in a text file named `students.txt`.  
- Uses `with open()` for safe reading and writing of data.  

## How to Run the Program
1. Open Jupyter Notebook  
2. Open `Student_Record_Management_System.ipynb`  
3. Run all cells or the main program cell  
4. Follow the menu displayed on the screen  

## Input Guidelines
- Student ID, Semester, and Marks must be numeric values  
- Student Name and Course should be text values  
- Menu options must be selected using numbers only  
- Invalid inputs will display appropriate error messages  

## Output
- Displays student records in a structured format  
- Confirms add, search, and delete operations  
- Saves data automatically before exiting  

## Files Included
- `Student_Record_Management_System.ipynb` – Main program file  
- `README.md` – Project documentation  
- `students.txt` – Data file (auto-generated)  

## Conclusion
The Student Record Management System successfully demonstrates the use of Python programming concepts to solve a real-world problem.
The project is simple, user-friendly, and can be extended further by adding features such as updating marks or generating reports.

## Author Details
- Name: Anusha DS  
- USN: 1BG24BA013  
- Course:MBA 
- Python Programming  
- Project name: Student Record Management System
```
