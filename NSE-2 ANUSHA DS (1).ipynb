```python
#Data Cleaning
#1. Fix Datatype (InvoiceDate)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
#2. Handle Missing Values – Drop
df.dropna(subset=['CustomerID'], inplace=True)
df
```

```python
#3. Handle Missing Values – Fill
df.loc[:, 'Description'] = df['Description'].fillna('Unknown')
df
```

```python
#4. Remove Duplicates
df = df.copy()

df.loc[:, 'Description'] = df['Description'].fillna('Unknown')
df = df.drop_duplicates()
df
```

```python
#5. Text Cleaning using str.strip() & casing
df['Description'] = df['Description'].str.strip().str.title()
df['Country'] = df['Country'].str.strip().str.upper()
df
```

```python
#6. Remove Invalid Values
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df
```

```python
#Create 3 New Columns
#1. Total Amount
df = df.copy()

df.loc[:, 'TotalAmount'] = df['Quantity'] * df['UnitPrice']
df
```

```python
#2. Order Month
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
df['OrderMonth'] = df['InvoiceDate'].dt.month
df
```

```python
#3. Conditional Column – Order Size
df = df.copy()
df['OrderType'] = np.where(df['Quantity'] >= 10, 'Bulk', 'Regular')
df
```

```python
#Insights Generation (5 Insights)
#Insight 1: Total Sales by Country (groupby + agg) 
df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False).head()
```

```python
#Insight 2: Top 5 Best-Selling Products (sort_values) 
df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head()
```

```python
#Insight 3: Most Frequent Order Type (value_counts)
df['OrderType'].value_counts()
```

```python
#df['OrderType'].value_counts()
pd.pivot_table(
    df,
    values='TotalAmount',
    index='OrderMonth',
    aggfunc='sum'
)

```

```python
#Insight 5: High-Value Transactions (Boolean Filtering)
df[df['TotalAmount'] > 1000][['InvoiceNo', 'Country', 'TotalAmount']].head()
```
